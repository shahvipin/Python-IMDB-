```python
df1=pd.DataFrame({'Reg_id':[1,2,3,4],'Name':['Andrew','Bob','Charlie','David']})
df1
```

```python
df2=pd.DataFrame({'Log_id':[1,2,3,4],'Name':['Xavier','Andrew','Yauren','Bob']})
df2
```

```python
pd.merge(df1,df2,on='Name',how='inner')
```

```python
#find the loing id all those pepole who has registered

left_join=pd.merge(df1,df2,on='Name',how='left')
left_join
```

```python
rig_join=pd.merge(df1,df2,on='Name',how='right')
rig_join
```

```python
#find out name of those people who have not attend any event

pd.merge(df1,df2,on='Name',how='outer')
```

```python
left_join[left_join['Reg_id'].isna()]
```

```python
left_join[left_join['Log_id'].isna()]['Name'].to_list()
```

```python
#find out the name of those people who attend the event but not registered

pd.merge(df1,df2,on='Name',how='outer',indicator=True)
```
